### b. We now implement a probabilistic model using conditional probability distributions for different weather-related events. The model includes variables such as season, temperature, rain, coat, and umbrella, and it defines the probabilities of these variables based on different conditions.

```python
# Conditional probabilities
P_Seasons = {'Spring': 0.25, 'Summer': 0.25, 'Autumn': 0.25, 'Winter': 0.25}
P_Temperature_given_Seasons = {
    'Spring': {'Cold': 0.4, 'Mild': 0.4, 'Hot': 0.2},
    'Summer': {'Cold': 0.2, 'Mild': 0.5, 'Hot': 0.3},
    'Autumn': {'Cold': 0.3, 'Mild': 0.4, 'Hot': 0.3},
    'Winter': {'Cold': 0.5, 'Mild': 0.3, 'Hot': 0.2}
}
P_Rain_given_Seasons_Temperature = {
    'Spring': {'Cold': {'No rain': 0.7, 'Rain': 0.3}, 'Mild': {'No rain': 0.6, 'Rain': 0.4}, 'Hot': {'No rain': 0.8, 'Rain': 0.2}},
    'Summer': {'Cold': {'No rain': 0.8, 'Rain': 0.2}, 'Mild': {'No rain': 0.9, 'Rain': 0.1}, 'Hot': {'No rain': 0.7, 'Rain': 0.3}},
    'Autumn': {'Cold': {'No rain': 0.6, 'Rain': 0.4}, 'Mild': {'No rain': 0.5, 'Rain': 0.5}, 'Hot': {'No rain': 0.7, 'Rain': 0.3}},
    'Winter': {'Cold': {'No rain': 0.7, 'Rain': 0.3}, 'Mild': {'No rain': 0.8, 'Rain': 0.2}, 'Hot': {'No rain': 0.6, 'Rain': 0.4}}
}
P_Coat_given_Seasons_Temperature_Rain = {
    'Spring': {'Cold': {'No rain': {'No coat': 0.45, 'Coat': 0.55}, 'Rain': {'No coat': 0.6, 'Coat': 0.4}},
               'Mild': {'No rain': {'No coat': 0.55, 'Coat': 0.45}, 'Rain': {'No coat': 0.5, 'Coat': 0.5}},
               'Hot': {'No rain': {'No coat': 0.7, 'Coat': 0.3}, 'Rain': {'No coat': 0.8, 'Coat': 0.2}}},
    'Summer': {'Cold': {'No rain': {'No coat': 0.6, 'Coat': 0.4}, 'Rain': {'No coat': 0.7, 'Coat': 0.3}},
               'Mild': {'No rain': {'No coat': 0.5, 'Coat': 0.5}, 'Rain': {'No coat': 0.4, 'Coat': 0.6}},
               'Hot': {'No rain': {'No coat': 0.3, 'Coat': 0.7}, 'Rain': {'No coat': 0.2, 'Coat': 0.8}}},
    'Autumn': {'Cold': {'No rain': {'No coat': 0.55, 'Coat': 0.45}, 'Rain': {'No coat': 0.6, 'Coat': 0.4}},
               'Mild': {'No rain': {'No coat': 0.6, 'Coat': 0.4}, 'Rain': {'No coat': 0.5, 'Coat': 0.5}},
               'Hot': {'No rain': {'No coat': 0.7, 'Coat': 0.3}, 'Rain': {'No coat': 0.8, 'Coat': 0.2}}},
    'Winter': {'Cold': {'No rain': {'No coat': 0.5, 'Coat': 0.5}, 'Rain': {'No coat': 0.6, 'Coat': 0.4}},
               'Mild': {'No rain': {'No coat': 0.4, 'Coat': 0.6}, 'Rain': {'No coat': 0.3, 'Coat': 0.7}},
               'Hot': {'No rain': {'No coat': 0.2, 'Coat': 0.8}, 'Rain': {'No coat': 0.1, 'Coat': 0.9}}}
}
P_Umbrella_given_Seasons_Coat = {
    'Spring': {'No coat': {'No umbrella': 0.7, 'Umbrella': 0.3}, 'Coat': {'No umbrella': 0.5, 'Umbrella': 0.5}},
    'Summer': {'No coat': {'No umbrella': 0.8, 'Umbrella': 0.2}, 'Coat': {'No umbrella': 0.6, 'Umbrella': 0.4}},
    'Autumn': {'No coat': {'No umbrella': 0.7, 'Umbrella': 0.3}, 'Coat': {'No umbrella': 0.6, 'Umbrella': 0.4}},
    'Winter': {'No coat': {'No umbrella': 0.8, 'Umbrella': 0.2}, 'Coat': {'No umbrella': 0.7, 'Umbrella': 0.3}}
}

# Function to calculate probability
def calculate_probability(season, temperature, rain, coat, umbrella):
    p_season = P_Seasons[season]
    p_temp_given_season = P_Temperature_given_Seasons[season][temperature]
    p_rain_given_season_temp = P_Rain_given_Seasons_Temperature[season][temperature][rain]
    p_coat_given_season_temp_rain = P_Coat_given_Seasons_Temperature_Rain[season][temperature][rain][coat]
    p_umbrella_given_season_coat = P_Umbrella_given_Seasons_Coat[season][coat][umbrella]

    # Calculate the joint probability
    joint_probability = p_season * p_temp_given_season * p_rain_given_season_temp * p_coat_given_season_temp_rain * p_umbrella_given_season_coat
    return joint_probability
```

### By inferring the probability of certain events or states, models can be used to make predictions about future occurrences.

```python
def calculate_inference_task_temperature_with_umbrella(umbrella):
    total_probability = 0
    for season in ['Spring', 'Summer', 'Autumn', 'Winter']:
        for temperature in ['Cold', 'Mild', 'Hot']:
            for rain in ['No rain', 'Rain']:
                for coat in ['No coat', 'Coat']:
                    total_probability += calculate_probability(season, temperature, rain, coat, umbrella)
    return total_probability

# Example
umbrella_example = 'Umbrella'

print(f"Inference Task: P(Temperature | Umbrella) with {umbrella_example}: {calculate_inference_task_temperature_with_umbrella(umbrella_example):.4f}")
```

### *Comparison between the inference probabilities:

### 1. The probability of carrying an umbrella in different temperatures is 0.3243 and actual rain in those temperatures is 0.300. This shows the correlation of there being rain and carrying of an umbrella.

###   However a small value of 0.0243 shows carrying umbrella even if there is no rain. This shows the probability of carrying umbrella when there is no rain in various temperatures.

```python
def calculate_probability(season, temperature, rain, coat, umbrella):
    p_season = P_Seasons[season]
    p_temp_given_season = P_Temperature_given_Seasons[season][temperature]
    p_rain_given_season_temp = P_Rain_given_Seasons_Temperature[season][temperature][rain]
    p_coat_given_season_temp_rain = P_Coat_given_Seasons_Temperature_Rain[season][temperature][rain][coat]
    p_umbrella_given_season_coat = P_Umbrella_given_Seasons_Coat[season][coat][umbrella]

    # Calculate the joint probability
    joint_probability = p_season * p_temp_given_season * p_rain_given_season_temp * p_coat_given_season_temp_rain * p_umbrella_given_season_coat
    return joint_probability

# Calculate all possible joint probability distributions
all_joint_probabilities = {}

for season in ['Spring', 'Summer', 'Autumn', 'Winter']:
    for temperature in ['Cold', 'Mild', 'Hot']:
        for rain in ['No rain', 'Rain']:
            for coat in ['No coat', 'Coat']:
                for umbrella in ['No umbrella', 'Umbrella']:
                    events = (season, temperature, rain, coat, umbrella)
                    probability = calculate_probability(*events)
                    all_joint_probabilities[events] = probability

# Print all joint probabilities
for events, probability in all_joint_probabilities.items():
    print(f"Joint Probability for {events}: {probability:.6f}")
```

### 1. Brute Force Inference

```python
def calculate_inference_task_season_with_hot_temperature_brute_force():
    total_probability = 0
    probabilities = []

    for season in ['Spring', 'Summer', 'Autumn', 'Winter']:
        for rain in ['No rain', 'Rain']:
            for coat in ['No coat', 'Coat']:
                for umbrella in ['No umbrella', 'Umbrella']:
                    joint_probability = calculate_probability(season, 'Hot', rain, coat, umbrella)
                    evidence_probability = calculate_probability(season, 'Hot', rain, 'No coat', 'No umbrella')
                    conditional_probability = joint_probability / evidence_probability
                    probabilities.append(conditional_probability)

    # Normalize the probabilities
    sum_probabilities = sum(probabilities)
    normalized_probabilities = [prob / sum_probabilities for prob in probabilities]

    # Print and return normalized probabilities
    index = 0
    for season in ['Spring', 'Summer', 'Autumn', 'Winter']:
        for rain in ['No rain', 'Rain']:
            for coat in ['No coat', 'Coat']:
                for umbrella in ['No umbrella', 'Umbrella']:
                    print(f"P({season} | Temperature = Hot, Rain = {rain}, Coat = {coat}, Umbrella = {umbrella}): {normalized_probabilities[index]:.4f}")
                    index += 1

    return sum(normalized_probabilities)

# Example
print("\nBrute Force Inference:")
print(f"Inference Task: P(Season | Temperature = Hot): {calculate_inference_task_season_with_hot_temperature_brute_force():.4f}\n")
```

```python
def calculate_inference_task_season_given_temperature_brute_force(temperature_condition, target_season='Summer'):
    total_probability = 0
    probabilities = []

    for season in ['Spring', 'Summer', 'Autumn', 'Winter']:
        for coat in ['No coat', 'Coat']:
            for umbrella in ['No umbrella', 'Umbrella']:
                # Provide a default value for 'rain'
                joint_probability = calculate_probability(season=season, temperature=temperature_condition, coat=coat, umbrella=umbrella, rain='No rain')
                evidence_probability = calculate_probability(season=season, temperature=temperature_condition, coat='No coat', umbrella='No umbrella', rain='No rain')
                conditional_probability = joint_probability / evidence_probability
                probabilities.append(conditional_probability)

    # Normalize the probabilities
    sum_probabilities = sum(probabilities)
    normalized_probabilities = [prob / sum_probabilities for prob in probabilities]

    # Print and return the probability for the target season
    index = 0
    for season in ['Spring', 'Summer', 'Autumn', 'Winter']:
        for coat in ['No coat', 'Coat']:
            for umbrella in ['No umbrella', 'Umbrella']:
                if season == target_season:
                    print(f"P(Season = {season} | Temperature = {temperature_condition}, Coat = {coat}, Umbrella = {umbrella}): {normalized_probabilities[index]:.4f}")
                    total_probability += normalized_probabilities[index]
                index += 1

    return total_probability

# Example
selected_temperature = 'Hot'
target_season = 'Summer'
print(f"\nBrute Force Inference for {target_season} given Temperature = {selected_temperature}:")
print(f"Inference Task: P({target_season} | Temperature = {selected_temperature}): {calculate_inference_task_season_given_temperature_brute_force(selected_temperature, target_season):.4f}\n")
```

### 2. Variable eliminiation inference

```python
def calculate_inference_task_season_with_hot_temperature_variable_elimination():
    temperature = 'Hot'
    seasons = ['Spring', 'Summer', 'Autumn', 'Winter']
    
    # Lists to store conditional probabilities
    conditional_probabilities = []

    for season in seasons:
        # Calculate joint probability
        joint_probability = calculate_probability(season, temperature, 'No rain', 'No coat', 'No umbrella')

        # Calculate evidence probability by summing over all possible values of rain, coat, and umbrella
        evidence_probability = 0
        for rain in ['No rain', 'Rain']:
            for coat in ['No coat', 'Coat']:
                for umbrella in ['No umbrella', 'Umbrella']:
                    evidence_probability += calculate_probability(season, temperature, rain, coat, umbrella)

        # Calculate conditional probability and append to the list
        conditional_probability = joint_probability / evidence_probability
        conditional_probabilities.append(conditional_probability)

    # Normalize the probabilities
    sum_conditional_probabilities = sum(conditional_probabilities)
    normalized_probabilities = [prob / sum_conditional_probabilities for prob in conditional_probabilities]

    # Rearrange probabilities based on the desired order
    sorted_probabilities = [normalized_probabilities[1], normalized_probabilities[0], normalized_probabilities[2], normalized_probabilities[3]]

    # Print normalized probabilities and inference task
    for season, sorted_probability in zip(['Autumn', 'Summer', 'Spring', 'Winter'], sorted_probabilities):
        print(f"P({season} | Temperature = {temperature}): {sorted_probability:.4f}")

    # Print the inference task
    target_season = 'Summer'
    target_probability = sorted_probabilities[seasons.index(target_season)]
    print(f"\nInference Task: P({target_season} | Temperature = {temperature}): {target_probability:.4f}")

    return sorted_probabilities

# Example
print("\nVariable Elimination Inference:")
calculate_inference_task_season_with_hot_temperature_variable_elimination()
```

### Variable elimination eliminates variables that are not part of the query, thereby reducing computational complexity.
### It is particularly useful for models with a certain structure as it avoids redundant computations and focuses on relevant parts of network.

### It is more scalable than brute force and can handle larger graphical models.

### Computational effort: 
### The computational effor depends on the size and structure of the graphical model. The computational complexity is reduced by eliminating irrelevant variables. It is efficient for graphical models with a certain structure. 

### Influence of choice of design:
### Since variable elimination is an algorithmic approach for exact inference in probabilistic graphical models, it eliminates irrelevant variables and computes exact probabilities. 
### It uses conditional independence relationships for reducing compuatational complexity. It may be more efficient than brute force for models with structured graphical patterns. The probablity of summer given the temperature being hot is found to be 0.3924.

### 3. Markov Chain Monte Carlo inference

```python
# Function to calculate probability for MCMC inference
def calculate_probability_mcmc(season, temperature):
    p_season = P_Seasons[season]
    p_temp_given_season = P_Temperature_given_Seasons[season][temperature]

    # Calculate the joint probability
    joint_probability = p_season * p_temp_given_season
    return joint_probability

# Update the MCMC inference function
def calculate_inference_task_season_with_hot_temperature_mcmc(num_samples=1000, burn_in=100, thinning=5):
    total_probability = 0
    temperature = 'Hot'

    for _ in range(num_samples + burn_in):
        # Sample from the joint distribution (season, temperature)
        sample_season = sample_from_distribution(['Spring', 'Summer', 'Autumn', 'Winter'])
        joint_probability = calculate_probability_mcmc(sample_season, temperature)

        if _ >= burn_in and (_ - burn_in) % thinning == 0:
            total_probability += joint_probability
            print(f"P({sample_season} | Temperature = {temperature}): {joint_probability:.4f}")

    return total_probability / (num_samples / thinning)

# Example
print("MCMC Inference:")
print(f"Inference Task: P(Season | Temperature = Hot): {calculate_inference_task_season_with_hot_temperature_mcmc():.4f}")
```

### Markov Chain Monte Carlo is a simulation-based approach.It is used in complex probabilistic models for estimating probabilities and making inferences.

### It works by generating a sequence of samples from the target probability by constructing a Markov chain.



### Computational effort: 
### Computational effort depends on the convergence rate and the number of iterations required to obtain a representative sample. It is useful when exact computation of probabilities in infeasible.

### Influence of choice of design:

### MCMC yields results accompanied by a measure of uncertainty. The accuracy increases with the number of samples and is suitable for approximating complex posterior distributions. It provides solutions by sampling from posterior distribution. The quality of approximation improves with more samples. The probability season given temperature being hot is found to be 0.0621
